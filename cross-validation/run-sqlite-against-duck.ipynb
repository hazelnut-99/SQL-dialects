```python
import os
duck_suites_dir = "../test-suites-generated/duck-query-log"
```

```python
con = sqlite3.connect("tutorial4.db")
cur = con.cursor()
```

```python
files = os.listdir(duck_suites_dir)
```

```python
failed = []
success = 0
for file_path in files:
    file_path = duck_suites_dir + "/" + file_path
    queries = []
    with open(file_path, 'r') as file:
        file_contents = file.read()
        queries = file_contents.split(";")
        queries = [query.lstrip("\n") for query in queries]
        queries = [query for query in queries if query]
        for sql in queries:
            try:
                res = cur.execute(sql)
                success += 1
            except Exception as e:
                failed.append((sql, e))
    print(file_path)
```
