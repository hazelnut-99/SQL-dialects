```python
import os
```

```python
def list_files_recursive(root_dir):
    file_list = []
    for root, dirs, files in os.walk(root_dir):
        for file in files:
            file_path = os.path.join(root, file)
            if (file_path.endswith("test") or file_path.endswith("test-slow")) and os.path.isfile(file_path):  
                file_list.append(file_path)

    return file_list
```

```python
file_list = list_files_recursive("/Users/hongshu/Desktop/eth-master-study/2024-spring/automated-software-testing/projects/databases-source-code/duckdb/test/sql")
```

```python
for file_path in file_list:
    enable_query_log_string = "statement ok\nSET log_query_path='/tmp/duck-query-log/%s-%s.log'\n" % (file_path.split("/")[-2], file_path.split("/")[-1])
    print(enable_query_log_string)
    
    with open(file_path, 'r') as f_in:
        original_content = f_in.read()

    
    modified_content = enable_query_log_string + original_content

    # Write the modified content to the output file
    with open(file_path, 'w') as f_out:
        f_out.write(modified_content)

    print(f"Words added to '{file_path}' successfully.")
```
